```python

import docker
from confluent_kafka import Producer
 
 
kafka_conf = {
    'bootstrap.servers': 'localhost:9092',
}
producer = Producer(kafka_conf)
log_topic = 'quickstart'
 
def delivery_report(err, msg):
    if err is not None:
        print(f"Message delivery failed: {err}")
    else:
        print(f"Message delivered to {msg.topic()} [{msg.partition()}]")
 
def fetch_postgresql_logs_from_container():
    client = docker.from_env()
    container = client.containers.get('postgres_db')  
    logs = container.exec_run('cat ./var/lib/postgresql/data/pg_log/postgresql-2024-08-01_042856.log').output.decode('utf-8')
    return logs.splitlines()
 
def send_data_to_kafka(data):
    for log in data:
        producer.produce(log_topic, key='key', value=log, callback=delivery_report)
    producer.flush()
 
logs = fetch_postgresql_logs_from_container()
send_data_to_kafka(logs)
 
 
```
